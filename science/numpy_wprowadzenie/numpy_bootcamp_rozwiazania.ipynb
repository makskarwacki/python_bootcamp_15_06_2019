# Indeksowanie, wycinanie, iterowanie - Indexing, Slicing and Iterating

### Stwórz liste sześcianów liczb od 0 do 9

```python
a = np.arange(10) ** 3
a
```

### wybierz z listy zaznaczone elementy

[ 0, 1, **8**, **27**, **64**, 125, 216, 343, 512, 729]

    

```python
a[2:5]
```

[ **0**, 1, **8**, 27, **64**, 125, 216, 343, 512, 729]

```python
a[:6:2]
```

### przekształć by otrzymać wynik

[-1000, 1,  -1000, 27, -1000,   125,   216,   343,   512, 729]

```python
a[:6:2]= -1000
a
```

### Odwróć listę

```python
a[::-1]
```

### Sprawdź typ 

```python
a.dtype
```

### W pętli for wyprintuj dla każdego elementu pierwiastek sześcienny z elementy

```python
for i in a:
    print(i**(1/3.))
```

### Stwórz macierz 10x10 z kwadratami liczb od 1 do 100

```python
a = np.arange(100).reshape((10, 10)) ** 2
a
```

### Stwórz macierz 5x5 gdzie na krawędziach będą 0 a w środku 1

```python
b = np.zeros(shape=(5, 5))
b[1:-1, 1:-1] = 1
b
```

### Stwórz tabliczkę mnożenia liczb od 0 do 9 w postaci dwuwymiarowej macierzy

```python
c = np.arange(10) * np.arange(10).reshape((10, 1))
c
```

### Obrót o wskazaną liczbę stopni

```python
from scipy import ndimage
angle=73
data = ndimage.rotate(im2, angle=angle)
plt.imshow(data)
```

### Czarno biały filtr

```python
import numpy as np
data = im2

sum_array = data.sum(axis=2, keepdims=True)
sum_array = sum_array / np.amax(sum_array)
data = np.ones(3) * sum_array

plt.imshow(data)
```

```python
data.shape
```

### Rozmazanie

```python
data = ndimage.gaussian_filter(im2, sigma=22.5)
plt.imshow(data)
```

### Wyostrzanie


```python
# blurred_f = ndimage.gaussian_filter(im2, 1)

data = im2
filter_blurred_f = ndimage.gaussian_filter(data, 1)

alpha = 0.99
sharpened = data + alpha * (data - filter_blurred_f)

plt.imshow(sharpened)
```
